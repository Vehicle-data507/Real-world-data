```python
import os
import pandas as pd
from tqdm import tqdm  # 导入tqdm库

def check_increasing_in_all_files():
    # 获取当前工作目录
    current_directory = os.getcwd()

    # 获取当前目录下所有Excel文件
    excel_files = [f for f in os.listdir(current_directory) if f.endswith('.xlsx') or f.endswith('.xls')]

    if not excel_files:
        print("当前目录下没有找到Excel文件。")
        return

    # 跳过的文件名
    skip_file = '2023_24_Shenzhen.xlsx'  # 假设是一个.xlsx文件，确保扩展名正确

    # 遍历每个Excel文件，添加进度条
    for file in tqdm(excel_files, desc="Checking for Increasing Date and Mileage"):
        if file == skip_file:  # 跳过指定文件
            print(f"Skipping file {file}")
            continue
        
        print(f"\n正在检查文件: {file}")
        file_path = os.path.join(current_directory, file)

        try:
            # 读取Excel文件
            df = pd.read_excel(file_path)

            # 检查是否有足够的列
            if df.shape[1] < 4:
                print(f"文件 {file} 中没有足够的列，至少需要4列。")
                continue

            # 获取第一列和第四列的数据
            first_column = df.iloc[:, 0]
            fourth_column = df.iloc[:, 3]

            # 检查第一列是否递增
            is_first_column_increasing = first_column.is_monotonic_increasing

            # 检查第四列是否递增
            is_fourth_column_increasing = fourth_column.is_monotonic_increasing

            # 打印结果
            if is_first_column_increasing:
                print("第一列是递增的。")
            else:
                print("第一列不是递增的。")

            if is_fourth_column_increasing:
                print("第四列是递增的。")
            else:
                print("第四列不是递增的。")

        except Exception as e:
            print(f"无法读取文件 {file}: {e}")

if __name__ == "__main__":
    check_increasing_in_all_files()
```
